```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'C:\Users\CaoDat\Pictures\vlcsnap-2022-03-24-23h40m44s310.png')
area_handle = np.array([[741,807],[2628,810],[171,1741],[3206,1759]],dtype = np.float32)
(w,h) =  (1000,1840)#cm
area_dst =  np.array([[0,0],[w,0],[0,h],[w,h]], dtype=np.float32)
```

```python
matrix_homography, _ = cv2.findHomography(area_handle, area_dst,cv2.RANSAC,5.0)
output = cv2.warpPerspective(img, matrix_homography, (w,h))
cv2.imwrite(r'..\images\image_map\minimap.png',output)
plt.imshow(output)
```
